# Fairness Module Demo - 10/24/2024
In our current filesystem, we have three maps: two generated by our algorithm (one for polsby-popper, the other for efficiency gap), and one representing the current district plan.

In this demo, we will run an analysis on the Polsby-Popper map and compare it to the efficiency gap map as well as the current district map.

# Demo 1: Single Map Analysis

```python
from FairnessModule import full_analysis, compare_maps
from cleancurrent import get_curr_district_file
import geopandas as gpd

# load PP map in pandas geodataframe
pp_map = gpd.read_file("shapeFile/AZ_PP/az_Polsby-Popper_districts2.shp")
pp_map.District = pp_map.District + 1


# perform fairness analysis on map
full_analysis(pp_map, verbose=True, showMap=True)
```

# Demo 2: Compare Maps

```python
# load the efficiency gap map
eg_map = gpd.read_file("shapeFile/AZ_EG/az_Efficiency-Gap_districts2.shp")
eg_map.District = eg_map.District + 1

compare_maps(pp_map, eg_map, showMaps=True)
```

```python
curr_map = get_curr_district_file('az')
curr_map.District += 1

compare_maps(pp_map, curr_map, showMaps=True)
```
