```python
from fairymander.generator import DistrictGenerator

my_generator = DistrictGenerator("az", 0.008, 100, 3, "competitiveness")

districts = my_generator.run_and_save(directory="my_districts", file_prefix="testing-az")
"""
Running this will save as the following file structure, creating the directory if it doesnt exist:

my_districts
    |
    |-testing_az
        |
        |-testing_az-0
        |-testing_az-1
        |-testing_az-2

Where the final "prefix-index" files will have all the .shp related files
"""
```

```python
from fairymander.fairness import full_analysis, compare_maps

full_analysis(districts[0])
```

```python
compare_maps(districts[0], districts[1])
```
